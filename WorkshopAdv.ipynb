Inital data

```python
from datetime import datetime

df_group = df.copy()
df_group['OrderDate'] = pd.to_datetime(df_group['OrderDate'], format='%Y-%m-%d')
df_group["Month"] = df_group.OrderDate.dt.month

df_group

df_pivot = df_group.pivot_table(index=["Month"], values=["Units"], aggfunc="sum")

df_pivot
```

GUI and Graph

```python
import ipywidgets as widgets
import matplotlib.pyplot as plt
from ipywidgets.widgets.widget_button import Button
from IPython.display import display
from IPython.display import clear_output

df_pivot2 = df_pivot.reset_index()
min_value = df_pivot2.Units.min()
max_value = df_pivot2.Units.max()

def potGraph(xValue, yValue, fValue):
  df_pivot3 = df_pivot2.copy()
  df_pivot3 = df_pivot3[df_pivot3.Units <= fValue]
  plt.plot(df_pivot3.Month, df_pivot3.Units)
  plt.xlabel(xValue)
  plt.ylabel(yValue)
  plt.show()

X_Asix = widgets.Text(
    value='',
    placeholder='Type X_Asix name',
    description='X_Asix:',
    disabled=False
)

Y_Asix = widgets.Text(
    value='',
    placeholder='Type Y_Asix name',
    description='Y_Asix:',
    disabled=False
)

Filter_value = widgets.IntSlider(
    value=max_value,
    min=min_value,
    max=max_value,
    step=10,
    description='Filter_value:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)


button = widgets.Button(description="Calculate!")
output = widgets.Output()
def on_button_clicked(b):
  with output:
    clear_output(True)
    potGraph(X_Asix.value, Y_Asix.value, Filter_value.value)

button.on_click(on_button_clicked)

display(X_Asix)
display(Y_Asix)
display(Filter_value)
display(button, output)


```
