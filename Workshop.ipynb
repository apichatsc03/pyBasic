```python
def convertYear(df):
  for i in range(0,len(df.index)):
    od = str(df.iloc[i,df.columns.get_loc('OrderDate')])
    ny = int(str(df.iloc[i,df.columns.get_loc('OrderDate')])[0:4]) + 543

    df.iloc[i,df.columns.get_loc('OrderDate')] = str(ny) + str(df.iloc[i,df.columns.get_loc('OrderDate')])[4:]
  return df

df = convertYear(df)
```

```python
def RoundUpPrice(df):
  for i in range(0,len(df.index)):
    op = df.iloc[i,df.columns.get_loc('Unit Price')]
    np = int(df.iloc[i,df.columns.get_loc('Unit Price')])
    if ((op-np) > 0.5):
      df.iloc[i,df.columns.get_loc('Unit Price')] = (np + 1)
    else:
      df.iloc[i,df.columns.get_loc('Unit Price')] = np
  return df

df = RoundUpPrice(df)
df
```

```python
df['Total sale']=""

def CalTotal(df):
  for i in range(0,43):
    df.iloc[i,df.columns.get_loc('Total sale')]=df.iloc[i,df.columns.get_loc('Units')]*df.iloc[i,df.columns.get_loc('Unit Price')]
  return df

df = CalTotal(df)

df
```
